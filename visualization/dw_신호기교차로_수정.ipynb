```python
import pandas as pd
```

```python
# 현재 디렉토리에서 상위 dataset 폴더 안의 엑셀파일
df = pd.read_excel('/Users/kyuree/Downloads/교차로 경위도.xlsx')
```

```python
# df파일의 상위 5개 행 확인
df.head()
```

```python
# df 사이즈 확인
df.shape
```

```python
# 필요없는 컬럼 제거 후 데이터프레임 새로 생성
df_streets = df[['index', '경도', '위도']]
df_streets.set_index('index', inplace=True)

#index 이름 제거
df_streets=df_streets.rename_axis(None)

# 컬럼 이름 변경
df_streets = df_streets.rename(columns={'위도': '경도', '경도': '위도'})
```

```python
df_streets
```

```python
import folium

# 파일 열었을 때 지도의 첫 화면 지정
streets_map1 = folium.Map(location=[37.566535, 126.9779692], zoom_start=6)
# nowonstreets_map
```

```python
# 데이터프레임의 행을 반복하며 마커 추가
for i in range(len(df_streets)):
    folium.Marker([df_streets.iloc[i][0], df_streets.iloc[i][1]], popup=df_streets.index[i]).add_to(streets_map1)
# for index, row in df_streets.iterrows():
#     folium.Marker([row['위도'], row['경도']], popup=index).add_to(nowonstreets_map)
```

```python
# 지도 저장 후 출력
streets_map1.save('송파구_중구_종로구map.html')
streets_map1
```
