```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
# 엑셀 파일 불러오기
excel_path1 = '/Users/kyuree/Desktop/pds_capstone1/dataset/교통사고2014_2016.xlsx'
excel_path2 ='/Users/kyuree/Desktop/pds_capstone1/dataset/교통사고2017_2019.xlsx'
excel_path3='/Users/kyuree/Desktop/pds_capstone1/dataset/교통사고2020_2022.xlsx'

# 각 시트를 데이터프레임으로 읽어와 리스트에 저장
dfs1 =pd.read_excel(excel_path1)
dfs2 =pd.read_excel(excel_path2)
dfs3 =pd.read_excel(excel_path3) 
# 시트별 데이터프레임을 행 기준으로 통합
accident = pd.concat([dfs1,dfs2,dfs3], axis=0, ignore_index=True)
# 결과 출력\n",
# accident
```

```python
#1. 필요한 컬럼만 필터링 
#'사고일시', '시군구', '사고내용', '사망자수', '중상자수', '경상자수', '부상신고자수', '사고유형','법규위반','도로형태', '가해운전자 상해정도','피해운전자 상해정도'\n",
filtered_accident = accident[['사고일시', '시군구', '사고내용', '사망자수', '중상자수', '경상자수', '부상신고자수', '사고유형','법규위반','도로형태', '가해운전자 상해정도','피해운전자 상해정도']]
# filtered_accident
```

```python
#사고일시 -> 사고발생년도 / 월 / 일 / 시간으로 각각 구분하는 컬럼으로 재구성\n",
for i in range(len(filtered_accident)):
    filtered_accident.loc[i,'사고발생_년도'] = filtered_accident.loc[i, '사고일시'].split()[0]
    filtered_accident.loc[i,'사고발생_월'] = filtered_accident.loc[i, '사고일시'].split()[1]
    filtered_accident.loc[i,'사고발생_일'] = filtered_accident.loc[i, '사고일시'].split()[2]
    filtered_accident.loc[i,'사고발생_시간'] = filtered_accident.loc[i, '사고일시'].split()[3]
# filtered_accident
```

```python
#확인
print(filtered_accident['사고발생_년도'].unique())
print(filtered_accident['사고발생_월'].unique())
print(filtered_accident['사고발생_일'].unique())
print(filtered_accident['사고발생_시간'].unique())
```

```python
#시군구 -> 서울특별시 /**구 / **동 으로 구분하는 컬럼 생성\n",
# print(filtered_accident['시군구'].value_counts())
for i in range(len(filtered_accident)):
    filtered_accident.loc[i,'자치구명']= filtered_accident.loc[i,'시군구'].split()[1]
    filtered_accident.loc[i,'행정동'] = filtered_accident.loc[i,'시군구'].split()[2]
# print(filtered_accident[['시군구','자치구명','행정동']])
# print(filtered_accident['행정동'].unique())
```

```python
#사고유형 -> 차대사람 / 세부사고유형 -> 횡단중으로 컬럼 분리
filtered_accident['사고유형'].value_counts()
for i in range(len(filtered_accident)):
    filtered_accident.loc[i,'사고세부유형'] = filtered_accident.loc[i,'사고유형'][7:]
# filtered_accident[['사고유형','사고세부유형']]
```

```python
#널값 확인
print(filtered_accident.isnull().sum())
# print(filtered_accident['피해운전자 상해정도'].value_counts())
```

```python
#도로형태 -> 도로형태(단일로) / 세부도로형태(기타)로 분리, 큰도로형태에 도로형태 넣음
for i in range(len(filtered_accident)):
    filtered_accident.loc[i,'세부도로형태'] = filtered_accident.loc[i,'도로형태'].split()[2]
    filtered_accident.loc[i,'큰도로형태'] = filtered_accident.loc[i,'도로형태'].split()[0]
    filtered_accident[['도로형태','큰도로형태','세부도로형태']]
filtered_accident['세부도로형태'].value_counts()
```
