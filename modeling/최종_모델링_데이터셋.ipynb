```python
#각 변수 데이터셋 불러오기
#[독립변수]1. 자치구별 점멸 신호기 개수 , 2. 자치구별 2017~2022년도 분기별 총 발생 교통사고 건수, 3.자치구별 2017~2022년도 분기별 생활인구수 평균 4. [타겟값]자치구별 생활인구수 대비 발생한 교통사고 발생률
import pandas as pd
import numpy as np

# 두 개의 엑셀 파일을 읽기
file1 = pd.read_csv('dw_quarter_avg.csv')
file2 = pd.read_csv('kr_모델링데이터셋.csv')
file3 = pd.read_csv('자치구별_생활인구대비_교통사고발생률.csv')

# 'Unnamed: 0' 컬럼명을 '자치구명'으로 바꾸기
file2.rename(columns={'Unnamed: 0': '자치구명'}, inplace=True)

# '자치구명'과 '자치구명' 컬럼을 기준으로 파일 병합
merged = pd.merge(file1, file2,on='자치구명')
merged=pd.merge(merged, file3,on='자치구명')
merged
```

```python
#최종 합본 csv파일로 저장
merged.to_csv('최종모델링_데이터셋.csv',index=True)
```
